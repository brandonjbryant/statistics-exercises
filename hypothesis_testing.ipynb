### T- Testing Exercises
##### 1.
- Ace Realty wants to determine whether the average time it takes to sell homes is different for its two offices.
    - A sample of 40 sales from office ,#1 revealed a mean of 90 days and a standard deviation of 15 days.
    - A sample of 50 sales from office ,#2 revealed a mean of 100 days and a standard deviation of 20 days. 
- Use a .05 level of significance.

```python
null_hypothesis = "The average time it takes to sell homes is the same at both offices"
alternative_hypothesis = "The average time it takes to sell a home is the same at Office # 1 "
confidence_level = .95
a = 1 - confidence_level # call your shot before you run your test
```

```python
# A one-tailed test has the entire 5% of the alpha level in one tail (in either the left, or the right tail). 
#A two-tailed test splits your alpha level in half (as in the image to the left).
n1 = 40
mean1 = 90
sd1 = 15

n2 = 50
mean2 = 100
sd2 = 20

t, p = stats.ttest_ind_from_stats(mean1, sd1, n1, mean2, sd2, n2)

one_tailed_p = p / 2
one_tailed_p
```

```python
#visualize
viz = n1 + n2 - 2

dist = stats.t(viz)
x = np.linspace(-4, 4)
y = dist.pdf(x)

plt.figure(figsize=(12, 8))
plt.plot(x, y)
plt.vlines([-2, 2], 0, .3)
```

```python
if (p/2) < a and t > 0:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
    print("we don't have enough evidence to support the claim that", alternative_hypothesis)
```

##### 2.
- Load the mpg dataset and use it to answer the following questions:

####  Is there a difference in fuel-efficiency in cars from 2008 vs 1999?
    
 

```python
null_hypothesis = "There is no difference in fuel-efficiency in cars from 2008 vs 1999"
alternative_hypothesis = "The fuel efficiency in cars from 2008 is different than those from 1999" 
confidence_level = .95
a = 1 - confidence_level # call your shot before you run your test
```

```python
mpg = pydataset.data('mpg')
mpg.head(40)
```

```python
x1 = mpg[mpg.year == 1999].hwy
x2 = mpg[mpg.year == 2008].hwy

x1.var(), x2.var()
```

```python
t, p = stats.ttest_ind(x1, x2)
p
```

```python
if (p/2) < a and t > 0:
    print("We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
    print("we don't have enough evidence to support the claim that", alternative_hypothesis)
```

###  Are compact cars more fuel-efficient than the average car?

```python
null_hypothesis = "there is no difference in fuel-efficiency between compact cars and the average car"
alternative_hypothesis = "there is a difference in fuel-efficiency between compact cars and the average car "
confidence_level = .95
a = 1 - confidence_level # call your shot before you run your test
```

```python
com = mpg[mpg['class'] == 'compact'].hwy
avg = mpg.hwy.mean()

t, p = stats.ttest_1samp(com, avg)
print('t=', t)
print('p=', p)
```

```python
if (p/2) < a and t > 0:
    print("We reject the null hypothesis that", null_hypothesis)
else:
    print("We fail to reject the null hypothesis that", null_hypothesis)
    print("we don't have enough evidence to support the claim that", alternative_hypothesis)
```

##### Do manual cars get better gas mileage than automatic cars?

```python
null_hypothesis = "there is no difference in gas mileage between manual and automatic cars"
alternative_hypothesis = "there is a difference in gas mileage between manual and automatic cars"
confidence_level = .95
a = 1 - confidence_level # call your shot before you run your test
```

```python
auto_transmission = mpg.trans.str.startswith('auto')

x1 = mpg[auto_transmission].hwy
x2 = mpg[~ auto_transmission].hwy

x1.var(), x2.var()
```

```python
# (scipy.stats.ttest_ind)  We can use this test, if we observe two independent samples from the same or different population, 
# e.g. exam scores of boys and girls or of two ethnic groups
t, p = stats.ttest_ind(x2, x1)
print('t=', t)
print('p=', p)
```

```python
if (p/2) < a and t > 0:
    print("We reject the null hypothesis that", null_hypothesis)
else:
    print("We fail to reject the null hypothesis that", null_hypothesis)
    print("we don't have enough evidence to support the claim that", alternative_hypothesis)
```
